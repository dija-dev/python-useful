```python
import torch
from transformers import BertTokenizer, BertModel
```

```python
bert_model_name = "cl-tohoku/bert-base-japanese-whole-word-masking"
tokenizer = BertTokenizer.from_pretrained(bert_model_name)
model = BertModel.from_pretrained(bert_model_name).eval()
```

```python
if torch.cuda.is_available():
    model = model.to("cuda")
```

```python
inputs = tokenizer(
    text=["私の名前は田中太郎です。", "昨日、衆議院の議長が辞職した。"],
    max_length=512,
    padding="max_length",
    truncation=True,
    return_tensors="pt",
)
inputs
```

```python
with torch.no_grad():
    outputs = model(**inputs)
outputs
```
